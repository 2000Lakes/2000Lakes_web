# Get canton separate geojson info and saved in a format that can be used for fitBounds in google API

```python
import pandas as pd
```

# Seperate canton geojson

```python
import geojson
with open('cantons.geojson') as f:
    gj = geojson.load(f)
```

```python
gj['features'][0]['geometry']['coordinates']
```

```python
saved_geo = {}
for i in range(len(gj['features'])):
        k = gj['features'][i]['id']
        
        coords = gj['features'][i]['geometry']['coordinates']
        cood_array = []
        
        for i in range(len(coords)):
            for j in range(len(coords[i])):
                if (isinstance(coords[i][j][0],list)):
                    print(len(coords[i][j]))
                    for n in range(len(coords[i][j])):
                        cood_array.append([coords[i][j][n][1],coords[i][j][n][0]])
                else:
                    cood_array.append([coords[i][j][1],coords[i][j][0]])
        saved_geo[k] = cood_array
```

```python
saved_geo
```

```python
df = pd.DataFrame()
df['canton_id'] = saved_geo.keys()
df['coords'] = saved_geo.values()
```

```python
df
```

```python
df.to_csv('canton_coords.csv',index=False)
```

## Get only two points for bounding

```python
saved_latlng = {}
for i in range(len(gj['features'])):
        k = gj['features'][i]['id']
        saved_lat = []
        saved_lng = []
        coords = gj['features'][i]['geometry']['coordinates']
        
        for i in range(len(coords)):
            for j in range(len(coords[i])):
                if (isinstance(coords[i][j][0],list)):
                    print(len(coords[i][j]))
                    for n in range(len(coords[i][j])):
                        saved_lat.append(coords[i][j][n][1])
                        saved_lng.append(coords[i][j][n][0])
                else:
                    saved_lat.append(coords[i][j][1])
                    saved_lng.append(coords[i][j][0])
        saved_latlng[k] = [max(saved_lat), min(saved_lat), max(saved_lng), min(saved_lng)]
```

```python
saved_latlng
```

```python
df = pd.DataFrame()
df['canton_id'] = saved_latlng.keys()
df['coords'] = saved_latlng.values()
```

```python
df.to_csv('canton_coords.csv',index=False)
```
